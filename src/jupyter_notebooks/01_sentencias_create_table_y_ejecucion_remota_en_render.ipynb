```python
tabla_rol = """CREATE TABLE Rol (
        Id_Rol SERIAL PRIMARY KEY,
        Nombre VARCHAR(255) NOT NULL UNIQUE
    );"""

tabla_modalidad = """CREATE TABLE Modalidad (
    Id_Modalidad SERIAL PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL UNIQUE
);"""

tabla_campus = """CREATE TABLE Campus (
    Id_Campus SERIAL PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL UNIQUE
);"""

tabla_promocion = """CREATE TABLE Promocion (
    Id_Promocion SERIAL PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL UNIQUE
);"""

tabla_vertical = """CREATE TABLE Vertical (
    Id_Vertical SERIAL PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL UNIQUE
);"""

tabla_proyecto = """CREATE TABLE Proyecto (
    Id_Proyecto SERIAL PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL
);"""

tabla_profesor = """CREATE TABLE Profesor (
    Id_Profesor SERIAL PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL,
    -- Claves Foráneas
    Id_Promocion INT NOT NULL REFERENCES Promocion(Id_Promocion),
    Id_Vertical INT NOT NULL REFERENCES Vertical(Id_Vertical),
    Id_Campus INT NOT NULL REFERENCES Campus(Id_Campus),
    Id_Rol INT NOT NULL REFERENCES Rol(Id_Rol),
    Id_Modalidad INT NOT NULL REFERENCES Modalidad(Id_Modalidad)
);"""

tabla_alumno = """CREATE TABLE Alumno (
    Id_Alumno SERIAL PRIMARY KEY,
    Nombre VARCHAR(255) NOT NULL,
    Email VARCHAR(255) NOT NULL UNIQUE,
    Fecha_comienzo DATE,
    -- Claves Foráneas
    Id_Promocion INT NOT NULL REFERENCES Promocion(Id_Promocion),
    Id_Campus INT NOT NULL REFERENCES Campus(Id_Campus)
);"""

tabla_proyecto_alumno = """CREATE TABLE Proyecto_Alumno (
    Id_Alumno INT NOT NULL REFERENCES Alumno(Id_Alumno),
    Id_Proyecto INT NOT NULL REFERENCES Proyecto(Id_Proyecto),
    Calificacion VARCHAR(255),
    -- Definición de la clave primaria compuesta
    PRIMARY KEY (Id_Alumno, Id_Proyecto)
);"""

tablas_a_crear = [tabla_rol, tabla_vertical, tabla_campus, tabla_modalidad, tabla_promocion, tabla_proyecto, tabla_alumno, tabla_profesor, tabla_proyecto_alumno]

for sql_tabla in tablas_a_crear:
    cursor.execute(sql_tabla)

conn.commit()

```
