### Cargamos los CSVs en Dataframes

### Creamos las sentencias de inserción SQL para nuestro modelo entidad-relacion

```python
## Lista que contiene todas las sentencias de creación de registros:
sentencias_sql = []

# Proyectos
sentencias_sql.append(
        f"INSERT INTO Proyecto (Nombre) VALUES ('HLF'), ('EDA'), ('BBDD'), ('ML'), ('Deployment'), ('WebDev'), ('FrontEnd'), ('BackEnd'), ('React'), ('FullStack');"
    );


# Verticales
verticales = claustro.Vertical.unique()

for nombre_vertical in verticales:
    sentencias_sql.append(
        f"INSERT INTO Vertical (Nombre) VALUES ('{nombre_vertical}');"
    );


# Modalidades
modalidades = claustro.Modalidad.unique()

for nombre_modalidad in modalidades:
    sentencias_sql.append(
        f"INSERT INTO Modalidad (Nombre) VALUES ('{nombre_modalidad}');"
    );


# Promociones
promociones_clase_1 = clase_1.Promoción.unique()
promociones_clase_2 = clase_2.Promoción.unique()
promociones_clase_3 = clase_3.Promoción.unique()
promociones_clase_4 = clase_4.Promoción.unique()
promociones_claustro = claustro.Promocion.unique()

promociones_unicas_global = np.unique(
    np.concatenate(
        [
            promociones_clase_1, 
            promociones_clase_2, 
            promociones_clase_3, 
            promociones_clase_4, 
            promociones_claustro]))

for nombre_promocion in promociones_unicas_global:
    sentencias_sql.append(
        f"INSERT INTO Promocion (Nombre) VALUES ('{nombre_promocion}');"
    );


# Campus
campus_clase_1 = clase_1.Campus.unique()
campus_clase_2 = clase_2.Campus.unique()
campus_clase_3 = clase_3.Campus.unique()
campus_clase_4 = clase_4.Campus.unique()
campus_claustro = claustro.Campus.unique()

claustros_unicos_global = np.unique(
    np.concatenate(
        [
            campus_clase_1, 
            campus_clase_2, 
            campus_clase_3, 
            campus_clase_4, 
            campus_claustro]))

for nombre_campus in claustros_unicos_global:
    sentencias_sql.append(
        f"INSERT INTO Campus (Nombre) VALUES ('{nombre_campus}');"
    );

# Roles
roles = claustro.Rol.unique()

for nombre_rol in roles:
    sentencias_sql.append(
        f"INSERT INTO Rol (Nombre) VALUES ('{nombre_rol}');"
    );


# Alumnos
date_format = '%d/%m/%Y'

for alumno in clase_1.itertuples():

    datetime_obj = datetime.strptime(alumno.Fecha_comienzo, date_format)
    fecha_comienzo = datetime_obj.strftime('%Y-%m-%d')

    sentencias_sql.append(
        f"INSERT INTO Alumno (Nombre, Email, Fecha_Comienzo, Id_Promocion, Id_Campus) VALUES ('{alumno.Nombre}', '{alumno.Email}', '{fecha_comienzo}', (SELECT Id_Promocion FROM Promocion WHERE Nombre = '{alumno.Promoción}'), (SELECT Id_Campus FROM Campus WHERE Nombre = '{alumno.Campus}'));"
    );

for alumno in clase_2.itertuples():
    datetime_obj = datetime.strptime(alumno.Fecha_comienzo, date_format)
    fecha_comienzo = datetime_obj.strftime('%Y-%m-%d')

    sentencias_sql.append(
        f"INSERT INTO Alumno (Nombre, Email, Fecha_Comienzo, Id_Promocion, Id_Campus) VALUES ('{alumno.Nombre}', '{alumno.Email}', '{fecha_comienzo}', (SELECT Id_Promocion FROM Promocion WHERE Nombre = '{alumno.Promoción}'), (SELECT Id_Campus FROM Campus WHERE Nombre = '{alumno.Campus}'));"
    );

for alumno in clase_3.itertuples():
    datetime_obj = datetime.strptime(alumno.Fecha_comienzo, date_format)
    fecha_comienzo = datetime_obj.strftime('%Y-%m-%d')

    sentencias_sql.append(
        f"INSERT INTO Alumno (Nombre, Email, Fecha_Comienzo, Id_Promocion, Id_Campus) VALUES ('{alumno.Nombre}', '{alumno.Email}', '{fecha_comienzo}', (SELECT Id_Promocion FROM Promocion WHERE Nombre = '{alumno.Promoción}'), (SELECT Id_Campus FROM Campus WHERE Nombre = '{alumno.Campus}'));"
    );

for alumno in clase_4.itertuples():
    datetime_obj = datetime.strptime(alumno.Fecha_comienzo, date_format)
    fecha_comienzo = datetime_obj.strftime('%Y-%m-%d')

    sentencias_sql.append(
        f"INSERT INTO Alumno (Nombre, Email, Fecha_Comienzo, Id_Promocion, Id_Campus) VALUES ('{alumno.Nombre}', '{alumno.Email}', '{fecha_comienzo}', (SELECT Id_Promocion FROM Promocion WHERE Nombre = '{alumno.Promoción}'), (SELECT Id_Campus FROM Campus WHERE Nombre = '{alumno.Campus}'));"
    );

# Profesores
for profesor in claustro.itertuples():

    sentencias_sql.append(
        f"INSERT INTO Profesor (Nombre, Id_Promocion, Id_Vertical, Id_Campus, Id_Rol, Id_Modalidad) VALUES ('{profesor.Nombre}', (SELECT Id_Promocion FROM Promocion WHERE Nombre='{profesor.Promocion}'), (SELECT Id_Vertical FROM Vertical WHERE Nombre='{profesor.Vertical}'), (SELECT Id_Campus FROM Campus WHERE Nombre = '{profesor.Campus}'), (SELECT Id_Rol FROM Rol WHERE Nombre = '{profesor.Rol}'), (SELECT Id_Modalidad FROM Modalidad WHERE Nombre = '{profesor.Modalidad}'));"
    );

# Relacion Proyecto_Alumno

for alumno in clase_1.itertuples():
    sentencias_sql.extend([
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='HLF'), '{alumno.Proyecto_HLF}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='EDA'), '{alumno.Proyecto_EDA}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='BBDD'), '{alumno.Proyecto_BBDD}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='ML'), '{alumno.Proyecto_ML}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='Deployment'), '{alumno.Proyecto_Deployment}');",
    ]);

for alumno in clase_2.itertuples():
    sentencias_sql.extend([
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='HLF'), '{alumno.Proyecto_HLF}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='EDA'), '{alumno.Proyecto_EDA}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='BBDD'), '{alumno.Proyecto_BBDD}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='ML'), '{alumno.Proyecto_ML}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='Deployment'), '{alumno.Proyecto_Deployment}');",
    ]);

for alumno in clase_3.itertuples():
    sentencias_sql.extend([
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='WebDev'), '{alumno.Proyecto_WebDev}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='FrontEnd'), '{alumno.Proyecto_FrontEnd}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='BackEnd'), '{alumno.Proyecto_Backend}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='React'), '{alumno.Proyecto_React}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='FullStack'), '{alumno.Proyecto_FullSatck}');",
    ]);

for alumno in clase_4.itertuples():
    sentencias_sql.extend([
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='WebDev'), '{alumno.Proyecto_WebDev}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='FrontEnd'), '{alumno.Proyecto_FrontEnd}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='BackEnd'), '{alumno.Proyecto_Backend}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='React'), '{alumno.Proyecto_React}');",
        f"INSERT INTO Proyecto_Alumno (Id_Alumno, Id_Proyecto, Calificacion) VALUES ((SELECT Id_Alumno FROM Alumno WHERE Email='{alumno.Email}'), (SELECT Id_Proyecto FROM Proyecto WHERE Nombre='FullStack'), '{alumno.Proyecto_FullSatck}');",
    ]);

## Imprimir todas las sentencias
print(*sentencias_sql, sep='\n')
```
